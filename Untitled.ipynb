```python
words = open("names.txt","r").read().splitlines()
```

```python
s = list(set("".join(words)))
s.append(".")
s.sort()
print(len(s))
```

```python
itos = {i:j for i,j in zip(range(27),s)}
stoi = {i:j for i,j in zip(s,range(27))}
```

```python
itos
```

```python
stoi
```

```python
import torch
N = torch.ones(27,27)
for w in words:
    w = "." + w + "."
    for ch1,ch2 in zip(w,w[1:]):
        i = stoi[ch1]
        j = stoi[ch2]
        N[i,j]+=1
N
```

```python
N = N.int()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.figure(figsize=(16,16))
plt.imshow(N, cmap="Blues")
for i in range(27):
    for j in range(27):
        ch = itos[i]+itos[j]
        plt.text(i, j, ch,ha="center",va="bottom",color="gray")
        plt.text(i, j, N[i,j].item(),ha="center",va="top",color="gray")
```

```python
P = N.float()
P = P/P.sum(1,keepdim=True)
P
```

```python
g = torch.Generator().manual_seed(2154548)
ix = 0
for i in range(20):
    word = ''
    while True:
        p = P[ix]
        ix = torch.multinomial(p,num_samples = 1,generator=g).item()
        word+=itos[ix]
        if ix==0:
            break
    print(word)
```
